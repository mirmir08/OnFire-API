```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


frutas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/fruit_data_with_colors.csv')
frutas.head()
```

# Explorar variables mediante la visualización

```python
plt.figure(figsize=(5, 3))

frutas['fruit_name'].value_counts().plot(kind='bar')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize':(5,3)})
sns.boxplot(x='fruit_name', y='width', data=frutas, color='lightblue')
```

```python
sns.set_theme(rc={'figure.figsize':(5,3)})
sns.violinplot(x='fruit_name', y='width', data=frutas)
```
