```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
#crear un modelo dentro del modulo nn
class Modelo(nn.Module):
  #cuatro neuronas para cada tipo de flor
  #
    def __init__(self, entrada=4, c1=8, c2=9, salida=3):
        self.fc1 = nn.Linear(entrada, c1)
        self.fc2 = nn.Linear(c1, c2)
        self.out = nn.Linear(c2, salida)
```
